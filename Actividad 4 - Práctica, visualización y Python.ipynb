```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_edu = pd.read_excel('data_analitica.xlsx')
df_edu.dtypes
```

```python
df_edu
```

```python
df_edu['nivel_edu'].unique()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y='ingreso_mensual', data=df_edu)
plt.title('Distribución de Ingresos Mensuales por Género')
plt.xlabel('Género')
plt.ylabel('Ingresos Mensuales')
plt.show()
```

```python
plt.figure(figsize=(8,6)) #Se debe crear la figura y especificar el tamaño de esta
sns.scatterplot(x='anios_esc', y='ingreso_mensual', data=df_edu)
plt.title('Correlación entre Años de Escolaridad e Ingreso Mensual')
plt.ylabel('Ingreso Mensual')
plt.xlabel('Años de Escolaridad')
plt.show
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='nivel_edu', y='anios_esc', data=df_edu)
plt.title('Distribución de Años de Escolaridad y Nivel educativo')
plt.ylabel('Años de Escolaridad')
plt.xlabel('Nivel Educativo')
plt.show()
```

```python
conteo = df_edu.groupby(['sex', 'nivel_edu']).size().reset_index(name='count')
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y='count', hue='nivel_edu', data=conteo)
plt.title('Distribución de Ingresos Mensuales por Género')
plt.xlabel('Género')
plt.ylabel('Ingresos Mensuales')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxenplot(x='nivel_edu', y='anios_esc', data=df_edu)
plt.title('Distribución de Años de Escolaridad y Nivel educativo')
plt.ylabel('Años de Escolaridad')
plt.xlabel('Nivel Educativo')
plt.show()
```
